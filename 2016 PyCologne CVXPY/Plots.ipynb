```python
import numpy as np
import matplotlib.pyplot as pl
import cvxpy as cvx
```

```python
pl.xkcd()
x = np.linspace(0, 4, 500)

pl.figure(0, figsize=(3, 2))
ax = pl.subplot(111)
ax.set_axis_bgcolor('white')
ax.plot(x, (x - 2)**2 - 5, c='r')
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
ax.set_ylim(-5.2, -1.0)
ax.set_xticks(range(5))
ax.set_yticks(range(-5, 0))

```
